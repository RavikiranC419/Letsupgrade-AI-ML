# KNN Algorithm

```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn import neighbors
from sklearn.metrics import accuracy_score, confusion_matrix
```

```python
knn_data = pd.read_csv("train.csv")
```

```python
knn_data.head()
```

```python
knn_data = knn_data.drop(["Cabin"], axis=1)
```

```python
knn_data["Fare"] = np.round_(knn_data["Fare"])
```

```python
knn_data = knn_data.drop(["Name","Ticket","PassengerId"], axis=1)
```

```python
le = LabelEncoder()
```

```python
knn_data["Sex"] = le.fit_transform(knn_data["Sex"])
knn_data["Embarked"] = le.fit_transform(knn_data["Embarked"])
knn_data.head()
```

```python
y = knn_data["Pclass"]
x = knn_data.drop(["Pclass"], axis=1)
```

```python
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size = 0.3, random_state = 0)
```

```python
acc_scores = dict()
def KNN(k):
    knn = neighbors.KNeighborsClassifier(n_neighbors = k)
    print(f"Accuracy score based on k = {k} \n{round(knn.fit(X_train,y_train).score(X_test,y_test)*100,4)}\n")

    pred1 = knn.predict(X_test)
    conf = confusion_matrix(y_test, pred1)
    print(f"Confusion Matrix: \n{conf}")

    
    print("*************************************************************************************")
```

```python
for x in range (1,623):
    KNN(x)
```

# SVM

```python
from sklearn import svm
```

```python
svm_data = pd.read_csv("train.csv")
```

```python
svm_data.head()
```

```python
svm_data = svm_data.drop(["Name","Ticket","PassengerId","Cabin"], axis=1)
```

```python
svm_data["Fare"] = np.round_(svm_data["Fare"])
```

```python
svm_data["Sex"] = le.fit_transform(svm_data["Sex"])
svm_data["Embarked"] = le.fit_transform(svm_data["Embarked"])
svm_data.head()
```

```python
y = svm_data["Pclass"]
x = svm_data.drop(["Pclass"], axis=1)
```

```python
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size = 0.3, random_state = 0)
```

```python
clf = svm.SVC(gamma = 0.01, C = 100)

def SVM(x, y):
    X_train, X_test, y_train, y_test = train_test_split(x, y,test_size=0.3, random_state = 0)
    y_pred = clf.fit(X_train, y_train).predict(X_test)
    acc = round(accuracy_score(y_test, y_pred, normalize = True) * 100, 2)
    confm = confusion_matrix(y_test, y_pred)
    return [acc, confm]
```

```python
models = []
for enum, i in enumerate(svm_data.columns):
    if i == 'Age' or i == 'Fare':
        continue
    y = svm_data.iloc[:, enum:enum+1]
    x = svm_data.drop(columns = y.columns)
    models.append([SVM(x, y), y.columns[0]])
```

```python
for _ in models:
    print(f'{_[1]} vs {str(svm_data.drop([_[1]], axis=1).columns.to_list()).strip("[]")}\n')
    print(f'Accuracy: {_[0][0]}')
    print(f'Confusion Matrix: \n {_[0][1]}\n')
    print(f"Out of total 267 records only {np.trace(_[0][1])} are successfully classified, giving a model accuracy of\
          {_[0][0]}%.")
    print('\n-----------------------\n')
```
